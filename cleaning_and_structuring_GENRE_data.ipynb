```python
def genre(data):
    df1 = pd.DataFrame(columns = ['Release Year', 'Title', 'Origin/Ethnicity', 'Director', 'Cast','Genre', 'Wiki Page', 'Plot','Plot_cleanned', 'Genre_1','Genre_updated'])
    for i,row in data.iterrows():
        if len(row['Genre_1'].split(','))>1:
            for i in row['Genre_1'].split(','):
               # print(i)
                df=pd.DataFrame(row).T
                df['Genre_updated']=i
                df1=pd.concat([df1,df])
        else:
            df=pd.DataFrame(row).T
            df['Genre_updated']=df['Genre_1']
            df1=pd.concat([df1,df])
            
    return df1.reset_index(drop=True)


def genre_split_on_space(data):
    df1 = pd.DataFrame(columns = ['Release Year', 'Title', 'Origin/Ethnicity', 'Director', 'Cast','Genre', 'Wiki Page', 'Plot','Plot_cleanned','Genre_updated','Genre_updated_again'])
    for i,row in data.iterrows():
        if len(row['Genre_updated'].split())>1:
            for i in row['Genre_updated'].split():
               # print(i)
                df=pd.DataFrame(row).T
                df['Genre_updated_again']=i
                df1=pd.concat([df1,df])
        else:
            df=pd.DataFrame(row).T
            df['Genre_updated_again']=df['Genre_updated']
            df1=pd.concat([df1,df])
            
    return df1.reset_index(drop=True)
```
